```python
import pandas as pd
import numpy as np
```

```python
# reqading the file
books_df= pd.read_csv('book.csv',encoding='latin-1')
books_df
```

```python
# getting the top 5 books 
books_df.iloc[0:5]
```

```python
#number of unique users in the dataset
len(books_df['User.ID'].unique())
```

```python
# #number of different books in the dataset
len(books_df['Book.Title'].unique())
```

```python
user_books_df = books_df.pivot_table(index='User.ID', 
                        columns='Book.Title', 
                        values='Book.Rating')
```

```python
user_books_df
```

```python
#Impute those NaNs with 0 values
user_books_df.fillna(0, inplace=True)
user_books_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances( user_books_df.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = books_df['User.ID'].unique()
user_sim_df.columns = books_df['User.ID'].unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
