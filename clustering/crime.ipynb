```python
# importing the basic libraries
import pandas as pd
import matplotlib.pylab as plt
from sklearn.cluster import KMeans
import numpy as np
```

```python
# READING THE DATA
crime=pd.read_csv("crime_data.csv")
crime
```

```python
# normalizing the functions
def norm_func(i):
    x = (i-i.min()) / (i.max() - i.min())
    return (x)
```

```python
# getting the normalized values
df_norm = norm_func(crime.iloc[:,1:])
df_norm
```

```python
#plotting elbow curve to decide on the number of clusters
from sklearn.cluster import KMeans
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 11):
    clf = KMeans(n_clusters=i)
    clf.fit(df_norm)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 11), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()
```

```python
# developing dendrogram for verification
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
#p = np.array(df_norm) # converting into numpy array format 
z = linkage(df_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=0.,  # rotates the x axis labels
    #leaf_font_size=8.,  # font size for the x axis labels
)
plt.show()
```

```python
# choosing clusters=5 
clf = KMeans(n_clusters=5)
y_kmeans = clf.fit_predict(df_norm)
```

```python
y_kmeans
#clf.cluster_centers_
clf.labels_
```

```python
md=pd.Series(y_kmeans)  # converting numpy array into pandas series object 
crime['clust']=md # creating a  new column and assigning it to new column 
crime
```

```python
# obtaining the cluster attributes
crime.iloc[:,1:].groupby(crime.clust).mean()
```

```python
# plot of population vs murder rate
crime.plot(x="UrbanPop",y ="Murder",c=clf.labels_,kind="scatter",s=50 ,cmap=plt.cm.coolwarm) 
plt.title('Clusters using KMeans')
```
