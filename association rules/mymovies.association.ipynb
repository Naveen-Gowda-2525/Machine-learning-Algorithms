```python
# importing the basic libraries
import pandas as pd
import numpy as np
```

```python
# reading the file
dataset=pd.read_csv('my_movies.csv')
dataset
```

```python
# dropping the unwanted columns
dataset.drop(['V1','V2','V3','V4','V5'],inplace=True,axis=1)
dataset
```

```python
# installing mlxtend
!pip install mlxtend
```

```python
from mlxtend.frequent_patterns import apriori,association_rules
frequent_itemsets = apriori(dataset, min_support=0.005, max_len=5,use_colnames = True)
frequent_itemsets 
```

```python
frequent_itemsets.sort_values('support',ascending = False,inplace=True)
frequent_itemsets.sort_values
```

```python
# obtaining the antecedents and consequents
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.sort_values('lift',ascending = False).head(10)
```
